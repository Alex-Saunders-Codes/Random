```python
new_names = []
for i in kd_list:
    horse = "".join(filter(lambda x: not x.isdigit(), i))
    horse_final = horse.strip()
    new_names.append(horse_final)
```

```python
final_list = [i for i in new_names if len(i) > 1]
```

```python
total_string_len = len(''.join(final_list))
average_name_length = int(total_string_len/len(final_list))
optimal_name = []
for i in range(0,average_name_length):
    letter_list = [string[i] for string in final_list if len(string) > i]
    most_common_letter = max(set(letter_list), key=letter_list.count)
    optimal_name.append(most_common_letter)
''.join(optimal_name)
```
